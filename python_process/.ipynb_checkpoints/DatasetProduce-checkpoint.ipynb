# 生成训练集与测试集

```python
#import
import os
import time
import json
import argparse
import sys
import numpy as np
import scipy.io as scio
import random 
import cv2
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
import numpy as np
import h5py
from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler
```

```python
#define
def FramePosRead(PosPath,n_target):
    if(n_target==1):
        FramePos=scio.loadmat(PosPath)
        FramePos=np.array([FramePos['R'],FramePos['V']])
        FramePos=FramePos.reshape((2,FramePos.shape[2]))
        FramePos=FramePos.transpose(1,0)
        FramePos=list(FramePos)
    else:
        FramePos=scio.loadmat(PosPath)
        FramePos=np.array([FramePos['R_multi'],FramePos['V_multi']])#(2,27880,n_target)
        FramePos=FramePos.transpose(1,0,2)#(27880,2,n_target)
        FramePos=list(FramePos)
    return FramePos

def MTD_DataRead(MTD_DataPath):
    MTD_Data=[]

    for frameind in tqdm(range(0,len(os.listdir(MTD_DataPath)))):
#     for frameind in tqdm(range(0,1100)):

        if(True):
            data=scio.loadmat(MTD_DataPath+'/frame_%d.mat'%frameind)
            MTD_simu_frame=np.expand_dims(np.array(data['MTD_simu_0']),axis=0)#(155,288)->(1, 155, 288)
        
        else:
            data=h5py.File(MTD_DataPath+'/frame_%d.mat'%frameind) 
            MTD_simu_frame=np.expand_dims(np.array(data['MTD_simu_0']),axis=0)#(288,155)->(1, 288, 155)
            MTD_simu_frame=MTD_simu_frame.transpose(0,2,1)
            
        MTD_simu_frame=abs(MTD_simu_frame)
        MTD_Data.append(MTD_simu_frame)#[12000](1, 155, 288)
        
    return MTD_Data

def MTD_DataMaxPool_v(MTD_Data):
    for i in tqdm(range(len(MTD_Data))):
        MTD_frame=MTD_Data[i]
        MTD_frame_new=np.zeros((MTD_frame.shape[0],MTD_frame.shape[1]//2,MTD_frame.shape[2]))
        for j in range(MTD_Data[0].shape[2]):
            x=MTD_frame[0,:,j]
            x=MaxPool_1D(x)
            MTD_frame_new[0,:,j]=x
        MTD_Data[i]=MTD_frame_new
    return MTD_Data

def MaxPool_2D(img):
    #img二维
    new_n_col=img.shape[0]//2
    new_n_row=img.shape[1]//2
    img_new=np.zeros((new_n_col,new_n_row))
    for i in range(new_n_col):
        for j in range(new_n_row):
            Pool=img[i*2:i*2+2,j*2:j*2+2]
            max_pool=max(Pool.reshape(4))
            img_new[i,j]=max_pool
    return img_new

def MaxPool_1D(x):
    #x一维
    new_x_len=len(x)//2
    x_new=np.zeros((new_x_len))
    for i in range(new_x_len):
        Pool=x[i*2:i*2+2]
        max_pool=max(Pool)
        x_new[i]=max_pool
    return x_new

#MTD_Data变形为一个win_size一组
def FrameMTDWin(win_size,MTD_Data):
    frame_win=[]
    for n_win in range(len(MTD_Data)//win_size):
        frame_win.append([])
        for i in range(win_size):
            frame_win[n_win].append(MTD_Data[n_win*win_size+i])
        frame_win[n_win]=np.array(frame_win[n_win]).transpose(1,0,2,3)
    return frame_win

def FrameConfmapWin(win_size,frame_confmaps):
    frame_win=[]
    for n_win in range(len(frame_confmaps)//win_size):
        frame_win.append([])
        for i in range(win_size):
            frame_win[n_win].append(frame_confmaps[n_win*win_size+i])
            frame_confmaps[n_win*win_size+i]=[]
        #frame_win[n_win]=np.array([np.zeros(Confmap_size),frame_win[n_win]])+1e-30#(16, 384, 224)
        frame_win[n_win]=np.expand_dims(np.array(frame_win[n_win]),axis=0)
        
    #两个通道，0号是噪声全零，1号的有目标的通道
    return frame_win#424*(2, 16, 384, 224)

def GenerateConfmap(pos,Confmap_size,R_range,V_max):
    
    R_pos,V_pos=pos[0],pos[1]
    R_min,R_max=R_range[0],R_range[1]
    n_row,n_col=Confmap_size[0],Confmap_size[1]
    
    if((R_pos==0)&(V_pos==0)):
        confmap=np.zeros(Confmap_size) 
    else:
        R_col=int((R_pos-R_min)/(R_max-R_min)*n_col)
        V_row=int(n_row/2)+int(V_pos/V_max*(n_row/2))
        confmap=np.zeros(Confmap_size)
        cv2.circle(confmap, (R_col,V_row), 1 , 1, cv2.FILLED)
        confmap = gaussian_filter(confmap, 0.5)
    #np.save('1.npy',confmap)
    #plt.imshow(confmap)
    #print(R,V,R_col,V_row)
    return confmap

def GenerateConfmaps(FramePos,Confmap_size,R_max,V_max,n_target):
    if(n_target==1):
        frame_confmap=[]
        for i in range(len(FramePos)):
            pos=FramePos[i]
            confmap=GenerateConfmap(pos,Confmap_size,R_max,V_max)
            frame_confmap.append(confmap)
    else:
        frame_confmap=[]
        
        for i in range(len(FramePos)):
            pos=FramePos[i]#pos=(2,n_target)
            confmap_multi=0
            
            for j in range(n_target):
                confmap=GenerateConfmap(pos[:,j],Confmap_size,R_max,V_max)
                confmap_multi+=confmap
                
            frame_confmap.append(confmap_multi)
           
            
    return frame_confmap

#速度维高通滤波
def MTD_Data_Hfilter(MTD_Data,V_target_max,filtering):
   
    v=np.linspace(-V_target_max,V_target_max,len(MTD_Data[0][0]))
    
    if(filtering=='mean'):
        print('直接赋平均值')
        plt.figure()
        plt.plot(v,20*np.log10(MTD_Data[2000][0,:,10]))
        for i in range(len(MTD_Data)):
            MTD_Data[i][:,int(147/2-7):int(147/2+7),:]=np.mean(MTD_Data[i]/10)
            
        plt.plot(v,20*np.log10(MTD_Data[2000][0,:,10]))
        plt.xlim([-3,3])
        
    else:
#         u,sigma=0.3,0.6
        u,sigma=0.3,1.2
        y_gauss=1-np.exp(-((v-u)**2)/(2*(sigma**2)))
        y_gauss=y_gauss/max(y_gauss)

        a,b=5,0.5
        y_atan=np.arctan(abs((v-u)**a))**b;
        minmaxscaler=MinMaxScaler(feature_range=(np.mean(MTD_Data[500])/np.max(MTD_Data[500]), 1))
        y_atan=minmaxscaler.fit_transform(y_atan.reshape(len(y_atan),1)).squeeze()#归一化
        if(filtering=='gauss'):
            y=y_gauss
            print('filtering:','gauss')
        else:
            y=y_atan
            print('filtering:','atan')
        plt.plot(v,y)
        plt.xlim([-3,3])


        plt.figure()
        plt.plot(v,20*np.log10(MTD_Data[500][0,:,10]),'b')
        for i in range(len(MTD_Data)):
            for j in range(MTD_Data[0].shape[2]):
                x=MTD_Data[i][0,:,j]
                MTD_Data[i][0,:,j]=x*y

        plt.plot(v,20*np.log10(MTD_Data[500][0,:,10]),'r')
        plt.legend(['original','gauss'])
        plt.savefig('gauss_result.png',dpi=1000)
    return MTD_Data

def TrainTestClip(data,label,rate):
    data=list(data)
    label=list(label)
    # 随机打乱数据与标签
    randnum = 30
    random.seed(randnum)
    random.shuffle(data)
    random.seed(randnum)
    random.shuffle(label)
    
    # 按照1-rate：rate分训练集和测试集
    Xtrain = data[0:int(len(data) *(1-rate))]
    Xtest = data[int(len(data) * (1-rate)):]

    Ytrain=label[0:int(len(label) * (1-rate))]
    Ytest=label[int(len(label) * (1-rate)):]
    Xtrain,Xtest,Ytrain,Ytest=np.array(Xtrain),np.array(Xtest),np.array(Ytrain),np.array(Ytest)
    return Xtrain,Xtest,Ytrain,Ytest
```

```python
n_target=1
win_size = 4
Confmap_size = [147,334]
V_target_max = 20
R_range=[1000,3000]

SCR=-30
filtering='mean'
maxpool='’

MTD_DataPath = './matlab_process/matlab_dataset/RDmaps_SCR'+str(SCR)+'dB/'
PosPath = './matlab_process/matlab_dataset/label.mat'
save_path ='./dataset/MTD_'+str(SCR)+'dB'+maxpool+'/'+filtering+'/'
if(os.path.exists(save_path)==False): os.makedirs(save_path)
```

```python
#读取所有MTD数据
MTD_Data=MTD_DataRead(MTD_DataPath)#返回一个list, [12000](1, 155, 288)
print(len(MTD_Data),MTD_Data[0].shape)
```

```python
MTD_Data=MTD_Data_Hfilter(MTD_Data,V_target_max,filtering=filtering)
```

```python
#读取所有的坐标(R,V)
FramePos = FramePosRead(PosPath,n_target)# 27880*(R,V)
```

```python
#MTD数据变为一个win_size一组
data_win=FrameMTDWin(win_size,MTD_Data)# 424*(16, 2, 155, 217)
MTD_Data.clear()
```

```python
#生成所有的confmap
frame_confmaps=GenerateConfmaps(FramePos,Confmap_size,R_range,V_target_max,n_target)#6792*(h, w)

#confmaps数据变为一个win_size一组
confmaps_win=FrameConfmapWin(win_size,frame_confmaps)#424*(2, 16, h, w)
frame_confmaps.clear()
```

```python
# 按照7：3分训练集和测试集
data_win=np.array(data_win)
confmaps_win=np.array(confmaps_win)
print(data_win.shape,confmaps_win.shape)

#连续的25*4帧组合在一起，然后切分,因为连续的帧混在一起很相似
data_win=data_win.reshape(-1,25,1,4,Confmap_size[0],Confmap_size[1])
confmaps_win=confmaps_win.reshape(-1,25,1,4,Confmap_size[0],Confmap_size[1])
train_data,test_data,train_confmaps,test_confmaps=TrainTestClip(data_win,confmaps_win,0.3)

#连续的25*4帧展开
train_data=train_data.reshape(-1,1,4,Confmap_size[0],Confmap_size[1])
test_data=test_data.reshape(-1,1,4,Confmap_size[0],Confmap_size[1])
train_confmaps=train_confmaps.reshape(-1,1,4,Confmap_size[0],Confmap_size[1])
test_confmaps=test_confmaps.reshape(-1,1,4,Confmap_size[0],Confmap_size[1])
print(train_data.shape,test_data.shape,train_confmaps.shape,test_confmaps.shape)
```

```python
np.save(save_path+'train_data.npy', train_data)
np.save(save_path+'test_data.npy', test_data)
np.save(save_path+'train_confmaps.npy', train_confmaps)
np.save(save_path+'test_confmaps.npy', test_confmaps)
```
